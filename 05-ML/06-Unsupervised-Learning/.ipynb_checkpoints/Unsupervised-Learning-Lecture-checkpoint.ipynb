# __*Unsupervised Learning Lecture*__

## High dimensional datasets are at risk of being very sparse: most training instances are likely to be far away from each other.

## Of course, this also means that a new instance will likely be far away from any training instance, making predictions much less reliable than in lower dimensions, since they will be based on much larger extrapolations.

## In short, the more dimensions the training set has, the greater the risk of overfitting it.

## __*The Wine Dataset*__

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.datasets import load_wine

wine = load_wine(as_frame=True)
X = wine.data
y = wine.target
wine_features = X.columns

#⚠️ Data must be centered around its mean before applying PCA ⚠️
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X), columns=wine_features)
X
```

```python
y
```

```python
sns.heatmap(pd.DataFrame(X).corr(), 
            cmap='coolwarm', 
            linewidth=0.5, 
            linecolor="black", 
            annot=True,
            annot_kws={"fontsize":6});
```

### __*Compute the Principal Components*__

```python
from sklearn.decomposition import PCA
# rembember: we don't have Y
pca = PCA()
pca.fit(X)
```

```python
# Access our 13 PCs 
W = pca.components_
# → rows = PCs, columns = original features (weights = loadings)

# Print PCs as COLUMNS
W = pd.DataFrame(W.T,
                 index=wine_features,
                 columns=[f'PC{i}' for i in range(1, 14)])
# → transpose so each PC becomes a column
# → rows = original features
# → values = loadings (weights in the linear combination that forms each PC)

W
# → display loadings matrix: which features contribute to which PC
```

### __*Project our dataset into this new space of PCs*__

```python
# Project original data onto the principal components (get PC scores)
X_proj = pca.transform(X)

# Convert projection to a DataFrame with PC columns
X_proj = pd.DataFrame(X_proj, columns=[f'PC{i}' for i in range(1, 14)])

# Show the PC scores for each observation
X_proj

# PC for each of our wines
```

```python
# Correlation matrix of the projected PCs (should be ~0 off-diagonal if PCA worked)
sns.heatmap(X_proj.corr(), 
            cmap='coolwarm', 
            linewidth=0.5, 
            linecolor="black");
```

### __*Computational Proof*__

```python
# Computational check of matrix dimensions
W = pca.components_.T  # Transpose so PCs are columns (p x k)

print("Shape of W: ", W.shape)  # (p, k)
print("Shape of X:", X.shape)   # (m, p)
#This confirms the shapes match the projection formula
```

```python
# Verify that transform(X) == projection X @ W (numerically)
np.allclose(
    pca.transform(X),
    np.dot(X, W)
)
# This confirms the core PCA relationship
```

## __*[EXTRA SLIDES](https://docs.google.com/presentation/d/1ntRBvMS9L2uOQzxBJWftAnjASEyJmeHjbP6vs1wzX5Y/edit?usp=sharing)*__

### __*We can do it with NumPy*__

np.linalg.eig(M) computes the eig_vals and eig_vecs of M

Covariance Matrix = X.T dot X 

of shape (p,p) (if features are centered)

❗️ eig() decomposition can take very long

```python
# Compute covariance matrix
# WHY: PCA needs to know how features vary together to find the directions
cov_matrix = np.cov(X, rowvar=False)

# Eigen-decomposition of covariance matrix
# WHY: eigenvectors = those directions of maximum shared variance (PCs)
#      eigenvalues = how much variance each direction contains (ranking PCs)
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
```

```python
eig_vals
```

```python
eig_vecs
```

```python
# Show all 13 principal components (unranked)

# Put eigenvectors into a DataFrame
# to see the loadings (weights) showing how each original feature contributes to each principal component
W = pd.DataFrame(eig_vecs, 
                 index=wine_features,
                 columns=[f'PC{i}' for i in range(1, 14)])

# Display unranked PCs (before sorting by eigenvalues)
W
```

### __*PCs are ranked by order of importance*__

```python
# Estimate variance explained by each PC
X_proj.std()**2 / ((X.std()**2).sum()) 
# => PC variance tells how much of the original data's total variance is preserved along that component (similar idea to R²)
```

```python
# Sklearn provides it automatically
pca.explained_variance_ratio_
```

```python
pca.singular_values_
# The singular values are just another form of the eigenvalues.
# You are doing PCA manually → using covariance + eigenvalues/eigenvectors
# sklearn does PCA automatically → using SVD (more stable + faster)
# Singular values tell you how strong each PC is.
# Larger value = more important direction.
```

```python
# Plot variance explained by each PC
# WHY: to see how many PCs are worth keeping (importance drops after some point)
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Principal Component')
plt.ylabel('% explained variance')
```

### __*PCA for Dimensionality Reduction*__

```python
# Cumulative explained variance plot
# WHY: to decide how many PCs to keep (e.g., choose smallest k reaching ~95%)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.ylim(ymin=0)
plt.title('Cumulated share of explained variance')
plt.xlabel('# of principal components used')  # keep around 95%
```

```python
# minimal number of PCs needed to explain at least 75% of the variance.
# => Find smallest k such that cumulative variance ≥ 75%
(np.cumsum(pca.explained_variance_ratio_) < 0.75).sum() + 1
# It counts how many PCs you need before reaching 75%, then adds 1 to include the first PC that gets you to 75% coverage.
```

### __*Test Model Performance (with k=3 Dimensions)*__

```python
# Fit PCA keeping only 3 principal components
# WHY: reduce dimensionality while preserving most important variance
pca3 = PCA(n_components=3).fit(X)
# if you put 0.9 instead of 3 -> it will keep as many as needed to have 90% 

# Project data into the 3-PC space
# WHY: create a new dataset with only the top 3 informative features
X_proj3 = pd.DataFrame(pca3.fit_transform(X),
                       columns=['PC1', 'PC2', 'PC3'])

# Reduced 3D representation of the original data
X_proj3
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Train + evaluate logistic regression using only the first 3 PCs
# WHY: check if reduced-dimension data still gives good accuracy
print("accuracy 3 PCs")
print(cross_val_score(LogisticRegression(), X_proj3, y, cv=5).mean())

# Train + evaluate using all original features (baseline)
# WHY: compare against full-feature performance
print("\naccuracy all 13 initial features")
print(cross_val_score(LogisticRegression(), X, y, cv=5).mean())
# Using 3 PCs gives slightly lower accuracy than all 13 features, but still performs very well while greatly reducing dimensionality.
# We are gaining that BUT losing interpretability!
```

### __*Decompress*__

❓ Can you perfectly reconstruct X from X_proj3?

- Not if you kept k < 13 dimensions; information has been lost
- We can approximate X by reconstructing it with inverse_transform()

```python
# Reconstruct an approximation of the original data from the 3 PCs
# WHY: to see how much information is lost when reducing to 3 components
X_reconstructed = pd.DataFrame(pca3.inverse_transform(X_proj3))

# Restore original column names for clarity
X_reconstructed.columns = X.columns

# Check reconstructed data shape (should match original X)
X_reconstructed.shape
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
sns.heatmap(X)
plt.title("original data")
plt.subplot(1,2,2)
plt.title("reconstructed data")
sns.heatmap(X_reconstructed);

#The heatmaps show that reducing to 3 PCs preserves the main structure of the data but loses some detail.
```

```python
# Accuracy
from sklearn.metrics import accuracy_score

y_pred = pd.Series(km.labels_).map({2:0, 0:1, 1:2}) # WARNING: change this manually!
accuracy_score(y_pred, y)
```

### __*What is n_init about?*__

sets the number of initializations to perform. This is important because two runs can converge on different cluster assignments. The default behavior for the scikit-learn algorithm is to perform ten k-means runs and return the results of the one with the lowest inertia!!

### __*Choosing Hyperparameter K*__

```python
inertias = []
ks = range(1,10)

for k in ks:
    km_test = KMeans(n_clusters=k, n_init=10).fit(X)
    inertias.append(km_test.inertia_)

plt.plot(ks, inertias)
plt.xlabel('k cluster number')
```
